```python
# Importing the libraries
```

```python
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf, pandas as pd
from ta import add_all_ta_features
from ta.utils import dropna
```

```python
#Importing the dataset
```

```python
dataset = pd.read_csv('IBM.csv')
```

```python
#plotting the dataset
```

```python

dataset.plot()
```

```python
#adding more features using ta package
```

```python
dataset  = dropna(dataset)

mom_data = add_all_ta_features(dataset, open="Open", high="High", low="Low", close="Adj_Close", volume="Volume", fillna = True) # Substantiate data with momentum indicators
```

```python
mom_data
```

```python
# Normalizing the data
```

```python
# making a copy
normalized_data = mom_data.copy()
```

```python
#making a list of the column names
columns_list = list(normalized_data.columns.values)
```

```python
#filtering out the columns which we dont want to normalize
columns_to_normalize = []
for item in columns_list:
    if item != "Date":
        columns_to_normalize.append(item)
        
```

```python
# normalizing the data
col_max_values = []
for col in columns_to_normalize:
    col_min = normalized_data[col].min()
    col_max = normalized_data[col].max()
    col_abs_max = max(abs(col_min), abs(col_max))
    normalized_data[col] = normalized_data[col].apply(lambda x: x/col_abs_max)
    
# saving the col_abs_max value.    
    col_max_values.append(col_abs_max)
    
    
```

```python
col_max_vals_df = pd.DataFrame([col_max_values], columns = columns_to_normalize)
col_max_vals_df
```

```python
normalized_data
```

```python
#setting splitting ratio
split_ratio = [0.75, 0.15, 0.10]

#setting set sizes
train_set_end = int(normalized_data.shape[0]*split_ratio[0])
val_set_end = train_set_end+int(normalized_data.shape[0]*split_ratio[1])
test_set_end = val_set_end+int(normalized_data.shape[0]*split_ratio[2])
```

```python
#splitting into train, val and test set
train_set = normalized_data.iloc[0:train_set_end, :]
val_set = normalized_data.iloc[train_set_end:val_set_end,:]
test_set = normalized_data.iloc[val_set_end:, :]
```

```python
print(train_set.shape, val_set.shape, test_set.shape)
```

```python
train_set.to_csv("./data/train_set.csv",index = False)
```

```python
val_set.to_csv("./data/val_set.csv",index = False)
```

```python
test_set.to_csv("./data/test_set.csv",index = False)
```
